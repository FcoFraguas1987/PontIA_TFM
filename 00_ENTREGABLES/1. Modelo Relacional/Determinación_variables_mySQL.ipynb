# Determinación ed variables mySQL

Uno de los objetivos de la creación del modelo relacional es que no suponga un sobrecoste tener una base de datos tipo SQL. Esto implica, entre otras coasa que las variable sno tengan un tamaño superior al de sus necesiades (en el momento actual).

Para ello tomaremos el df_final_v1 y veremos el rango de los valores de las variables numéricas así como el tamaño máximo de las variables categóricas.

```python
import pandas as pd
import numpy as np
```

```python
pwd
```

```python
#cargamos el df
df=pd.read_csv("../3.Archivos_ejecutables/dataset/df_final_v1", index_col=0) #indicamos index_col=0 porque el csv ya tiene indices
df.head()
```

```python
#Estudiamos el df
df.info()
```

```python
df_filtrado = df.select_dtypes(include="str")
```

```python
variables_str=list(df_filtrado.columns)
```

```python
for variable in variables_str:
    print(variable, df_filtrado[variable].drop_duplicates().apply(lambda x: (len(x))).max())

#print(df_filtrado[[variables_str]].apply(lambda x: len(x)))
```

```python
df_filtrado_2=df.select_dtypes(exclude="str")
variables_num=list(df_filtrado_2.columns)
```

```python
for variable in variables_num:
    print(variable, df_filtrado_2[variable].drop_duplicates().max())
```
