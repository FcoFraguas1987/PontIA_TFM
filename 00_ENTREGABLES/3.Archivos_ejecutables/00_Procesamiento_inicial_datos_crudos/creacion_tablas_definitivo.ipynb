```python
# 1) Imports
import json
import pandas as pd
from pathlib import Path
```

```python
# DATASET INFO_LOTES
```

```python
# 2) Ruta
ruta = Path("..") / "dataset" / "info_lote.json"

# Verificación
print("Existe archivo?:", ruta.exists(), "| Ruta:", ruta)

# 3) Leer el JSON
with open(ruta, "r", encoding="utf-8") as f:
    data = json.load(f)

print("Tipo dato:", type(data))

rows = []
for lote, lista in data.items():
    for item in lista:
        item["lote"] = lote  # añadimos la clave como nueva columna
        rows.append(item)


df_info_lotes = pd.DataFrame(rows)
df_info_lotes.head(1)

```

```python
# DATASET LOTES
```

```python
# 2) Ruta
ruta_2 = Path("..") / "dataset" / "lotes.json"

# Verificación
print("Existe archivo?:", ruta_2.exists(), "| Ruta:", ruta_2)

# 3) Leer el JSON
with open(ruta_2, "r", encoding="utf-8") as f:
    data_2 = json.load(f)

print("Tipo dato:", type(data_2))

df_lotes= pd.DataFrame(list(data_2.items()), columns=["t_id", "lote"])
df_lotes.head(1)
```

```python
# DATASET PROVEEDORES
```

```python
# 2) Ruta
ruta_3 = Path("..") / "dataset" / "proveedores.json"

# Verificación
print("Existe archivo?:", ruta_3.exists(), "| Ruta:", ruta_3)

# 3) Leer el JSON
with open(ruta_3, "r", encoding="utf-8") as f:
    data_3 = json.load(f)

print("Tipo dato:", type(data_3))

# Creamos una lista con las columnas que necesitamos en el DF
rows_3 = []
for proveedor, lista in data_3.items():
    for item in lista:
        item["proveedor"] = proveedor  # añadimos la clave como nueva columna
        rows_3.append(item)


df_proveedores = pd.DataFrame(rows_3)
df_proveedores.head(1)
```

```python
# DATASET PROVEEDORES-CLIENTES
```

```python
# 2) Ruta
ruta_4 = Path("..") / "dataset" / "proveedores-clientes.json"

# Verificación
print("Existe archivo?:", ruta_4.exists(), "| Ruta:", ruta_4)

# 3) Leer el JSON
with open(ruta_4, "r", encoding="utf-8") as f:
    data_4 = json.load(f)

print("Tipo dato:", type(data_4))

# Creamos una lista para los nombres de las columnas del DF
rows_4 = []

for proveedor, lista in data_4.items():
    for item in lista:
        # item = ["image12.png", "CompraMaestra"] ejemplo de lógica al indexar
        t_id = item[0] if len(item) > 0 else None
        cliente = item[1] if len(item) > 1 else None

        rows_4.append({
            "proveedor": proveedor,
            "t_id": t_id,
            "cliente": cliente
        })


df_proveedores_clientes = pd.DataFrame(rows_4)
df_proveedores_clientes.head(1)
```

```python
# DATASET TIPO
```

```python
# 2) Ruta
ruta_5 = Path("..") / "dataset" / "tipo.json"

# Verificación
print("Existe archivo?:", ruta_5.exists(), "| Ruta:", ruta_5)

# 3) Leer el JSON
with open(ruta_5, "r", encoding="utf-8") as f:
    data_5 = json.load(f)

print("Tipo dato:", type(data_5))

# Creamos una lista con las columnas que necesitamos en el DF
rows_5 = []
for tipo, lista in data_5.items():
    for item in lista:
        item["tipo"] = tipo  # añadimos la clave como nueva columna
        rows_5.append(item)


df_tipo = pd.DataFrame(rows_5)
df_tipo.head(1)
```

```python
# DATASET VENTA EN CLIENTE
```

```python
# 2) Ruta
ruta_6 = Path("..") / "dataset" / "venta_en_cliente.json"

# Verificación
print("Existe archivo?:", ruta_6.exists(), "| Ruta:", ruta_6)

# 3) Leer el JSON
with open(ruta_6, "r", encoding="utf-8") as f:
    data_6 = json.load(f)

print("Tipo dato:", type(data_6))

# Creamos una lista con las columnas para el DF
rows_6 = []
for cliente, lista in data_6.items():
    for item in lista:
        item["cliente"] = cliente  # añadimos la clave como nueva columna
        rows_6.append(item)


df_venta_en_cliente = pd.DataFrame(rows_6)
df_venta_en_cliente.head(1)
```

TABLAS MAESTRAS

```python
# FRUTAS

frutas = df_tipo[['tipo']].drop_duplicates().reset_index(drop=True)
frutas['id_tipo'] = frutas.index + 1
frutas = frutas[['id_tipo', 'tipo']]
frutas
```

```python
# MARCA

marca = df_info_lotes[['marca']].drop_duplicates().reset_index(drop=True)
marca['id_marca'] = marca.index + 1
marca = marca[['id_marca', 'marca']]
marca
```

```python
# PROVEEDORES

proveedores = df_proveedores_clientes[['proveedor']].drop_duplicates().reset_index(drop=True)
proveedores['id_proveedor'] = proveedores.index + 1
proveedores = proveedores[['id_proveedor', 'proveedor']]
proveedores
```

```python
# CLIENTES

clientes = df_proveedores_clientes[['cliente']].drop_duplicates().reset_index(drop=True)
clientes['id_cliente'] = clientes.index + 1
clientes = clientes[['id_cliente', 'cliente']]
clientes
```

TABLAS DINÁMICAS

```python
# LOTES

lotes = df_lotes[['lote', 't_id']].drop_duplicates().reset_index(drop=True)
lotes['id_lote'] = lotes.index + 1
lotes = lotes[['id_lote', 'lote', 't_id']]
lotes
```

MERGE ENTRE DF_INFO_LOTES Y TABLA MARCA

```python
# Creamos estas lineas de código para hacerlo más robusto. Es decir, en caso de ejecutar más de una vez el merge, hacemos una limpieza inicial para evitar
# duplicados de la variable id_marca y asi no entra en error.
evitar_conflicto = [c for c in df_info_lotes.columns if c.startswith('id_marca')] # con list comprehension filtramos en las columnas que empiezan por id_marca
df_info_lotes = df_info_lotes.drop(columns=evitar_conflicto, errors='ignore')     # Machacamos el df_info_lotes indicando la eliminación de las columnas del df anterior
                                                                                  # mediante errors para igonarar si alguna columna no existe

# merge limpio
df_info_lotes = df_info_lotes.merge(
    marca[['id_marca', 'marca']],
    on='marca',
    how='left',
    validate='many_to_one'  # comprobar que no haya duplicados en la tabla de la derecha
)

df_info_lotes
```

NUEVO DF DE INFO_LOTES DONDE ME QUEDO CON LOS VALORES UNICOS DE t_id Y id_marca

```python
df_info_lotes_unique = (
    df_info_lotes[['t_id', 'id_marca']].drop_duplicates(subset=['t_id'])
)
df_info_lotes_unique['t_id'].duplicated().sum()
df_info_lotes_unique
```

MERGE ENTRE DF LOTES Y EL NUEVO DF_INFO_LOTES_UNIQUE

```python
evitar_conflicto_2 = [c for c in lotes.columns if c.startswith('id_marca')]
lotes = lotes.drop(columns=evitar_conflicto_2, errors='ignore')

lotes = lotes.merge(
    df_info_lotes_unique,
    on='t_id',
    how='left',
    validate='many_to_one'
)

lotes
```

MERGE ENTRE DF_TIPO Y  TABLAS FRUTAS

```python
evitar_conflicto_3 = [c for c in df_tipo.columns if c.startswith('id_tipo')]
df_tipo = df_tipo.drop(columns=evitar_conflicto_3, errors='ignore')

df_tipo = df_tipo.merge(
    frutas[['id_tipo', 'tipo']],
    on='tipo',
    how='left',
    validate='many_to_one'
)

df_tipo
```

VALORES UNICOS DE DF_TIPO EN LAS VARIABLES t_id Y id_tipo

```python
df_tipo_unique = (
    df_tipo[['t_id', 'id_tipo']].drop_duplicates(subset=['t_id'])
)
df_tipo_unique['t_id'].duplicated().sum()
df_tipo_unique
```

MERGE ENTRE  LOTES Y DF_TIPO_UNIQUE

```python
evitar_conflicto_4 = [c for c in lotes.columns if c.startswith('id_tipo')]
lotes = lotes.drop(columns=evitar_conflicto_4, errors='ignore')

lotes = lotes.merge(
    df_tipo_unique,
    on='t_id',
    how='left',
    validate='many_to_one'
)

lotes
```

MERGE ENTRE DF_VENTA_EN_CLIENTE Y TABLA CLIENTES

```python
evitar_conflicto_10 = [c for c in df_venta_en_cliente.columns if c.startswith('id_cliente')]
df_venta_en_cliente = df_venta_en_cliente.drop(columns=evitar_conflicto_10, errors='ignore')

df_venta_en_cliente = df_venta_en_cliente.merge(
    clientes,
    on='cliente',
    how='left',
    validate='many_to_one'
)

df_venta_en_cliente
```
