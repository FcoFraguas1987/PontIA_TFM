## Preprocesamiento
En la carpeta donde estaban las imagenes se han generado dos archivos, uno con todos los documentos y otro sólo con las carpetas. Los archivos se obtenian con estos comandos

find . -type f > file_path
find . -type d > folder_path

Y a partir de aqui, se ha creado un archivo extrayendo del file_path el folder_path

grep -vxFf folder_path file_path > solo_archivos.txt

El proceso va a ser cargar un archivo que tiene la lista de los paths de las imagenes, y generar un dataset con pathb y frutas

```python
#nombre_columnas=["path"]
df_raw = pd.read_csv("solo_archivos.txt", header=None)
#df_raw.column=nombre_columnas
df_raw
```

```python
df_raw["fruta"]=df_raw[0].str.split("/").str[0]
df_raw
```

```python
#Calculamos las cantidades de imagenes por fruta y sus porcentajes
df_raw["fruta"].value_counts() #Esto son los valores correctos

df_raw["percent"]=(df_raw.groupby("fruta")["fruta"].transform("size")/len(df_raw))*100
```
